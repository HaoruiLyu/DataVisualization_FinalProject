This dataset contains flight data collected during the specified period (2022). It includes information on flight schedules, actual departure and arrival times, delays, cancellations and the reasons for these disruptions. The data can be used to analyze trends in flight on-time performance, identify common causes of delays and cancellations, and evaluate airline performance over time.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv(r"D:\PersonalFiles\IS445\FinalProject\data_2022.csv")
data
```

## Chart 1: Bar chart of Number of Flights per Airline

```python
# Count the number of flights for each airline and store it in 'airline_counts'
airline_counts = data['AIRLINE'].value_counts()

# plot the chart
plt.figure(figsize=(10, 6))
airline_counts.plot(kind='bar')
plt.title('Number of Flights per Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45) # Rotate the labels on the x-axis by 45 degrees

plt.show()
```

## Chart2: Bar chart of the TOP 10 cities by number of departing flights

```python
# Count the number of departing flights from each city and select the top 10 cities
airport_counts = data['ORIGIN_CITY'].value_counts().head(10)

# plot the chart
plt.figure(figsize=(12, 7))
airport_counts.plot(kind='bar')
plt.title('TOP 10 cities by number of departing flights')
plt.xlabel('City')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45) # Rotate the labels on the x-axis by 45 degrees

plt.show()
```

## Chart 3: Bar chart of Number of Cancelled Flights per Airline

```python
# Filter the dataset to include only the records where flights were cancelled
cancelled_flights = data[data['CANCELLED'] == 1]

# Count the number of cancelled flights for each airline
cancelled_counts = cancelled_flights['AIRLINE'].value_counts()

# plot the chart
plt.figure(figsize=(12, 7))
cancelled_counts.plot(kind='bar', color='red')
plt.title('Number of Cancelled Flights per Airline')
plt.xlabel('Airline')
plt.ylabel('Number of Cancelled Flights')
plt.xticks(rotation=45) # Rotate the labels on the x-axis by 45 degrees

plt.show()
```
