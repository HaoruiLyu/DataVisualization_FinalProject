```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv(r"D:\PersonalFiles\IS445\FinalProject\data_2022.csv")
data
```

## Chart 1: Pie chart of Percentage Distribution of Flight Delays by Reason

```python
# Filter out rows where all delay reasons are missing and select specific columns related to delay reasons
delay_data = data[['DELAY_DUE_CARRIER', 'DELAY_DUE_WEATHER', 'DELAY_DUE_NAS', 'DELAY_DUE_SECURITY', 'DELAY_DUE_LATE_AIRCRAFT']].dropna(how='all')

# Sum up each column to get the total delays for each reason
delay_summary = delay_data.sum()

# Define labels for the pie chart that are more readable
simplified_labels = ['Carrier', 'Weather', 'National Airspace System', 'Security', 'Late Aircraft']

# plot the chart
plt.figure(figsize=(8, 8))
plt.pie(delay_summary, labels=simplified_labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'orange', 'red', 'purple'])
plt.title('Percentage Distribution of Flight Delays by Reason')
plt.tight_layout() # Adjust the layout to ensure no part of the chart is cut off
plt.show()
```

## Chart2: Line and bar chart of Monthly Trends in Flight Delays and Cancellations

```python
# Convert 'FL_DATE' to a datetime type to facilitate resampling
data['FL_DATE'] = pd.to_datetime(data['FL_DATE'])

# Group the data by month and calculate the average delays and total cancellations for each month
monthly_data = data.groupby(data['FL_DATE'].dt.to_period("M")).agg({
    'DEP_DELAY': 'mean',   # Average departure delay
    'ARR_DELAY': 'mean',   # Average arrival delay
    'CANCELLED': 'sum'     # Total cancellations
}).reset_index()

monthly_data.rename(columns={'FL_DATE': 'Month'}, inplace=True)

# plot the chart
fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'tab:red'
ax1.set_xlabel('Month')
ax1.set_ylabel('Average Delay (minutes)', color=color)
ax1.plot(monthly_data['Month'].astype(str), monthly_data['DEP_DELAY'], label='Departure Delay', color='red')
ax1.plot(monthly_data['Month'].astype(str), monthly_data['ARR_DELAY'], label='Arrival Delay', color='blue')
ax1.tick_params(axis='y', labelcolor=color)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Total Cancellations', color=color)  
ax2.bar(monthly_data['Month'].astype(str), monthly_data['CANCELLED'], color='gray', alpha=0.6)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout() # Adjust the layout to prevent any clipping
plt.title('Monthly Trends in Flight Delays and Cancellations')
plt.show()
```

## Chart 3: Bar chart of Top 10 Busiest Flight Routes

```python
# Group data by origin and destination airport pairs and count the number of flights for each pair
route_data = data.groupby(['ORIGIN', 'DEST']).size().reset_index(name='Flight_Count')

# Sort the resulting DataFrame by 'Flight_Count' in descending order and take the top 10
route_data_sorted = route_data.sort_values(by='Flight_Count', ascending=False).head(10)

# plot the chart
plt.figure(figsize=(12, 8))
plt.barh(route_data_sorted['ORIGIN'] + ' - ' + route_data_sorted['DEST'], route_data_sorted['Flight_Count'], color='teal')
plt.xlabel('Number of Flights')
plt.ylabel('Route')
plt.title('Top 10 Busiest Flight Routes')
plt.gca().invert_yaxis() # Invert the y-axis to have the busiest route at the top of the chart
plt.tight_layout() # Adjust the layout to make sure there is no overlap in the plot elements
plt.show()
```
